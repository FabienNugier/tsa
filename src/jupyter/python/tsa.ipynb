```python
import numpy as np
def to_lstm_input(input, timesteps):
    result = np.empty((np.shape(input)[0] - timesteps + 1, timesteps, np.shape(input)[1]))
    for i in range(np.shape(input)[0] - timesteps + 1):
        result[i,:,:] = input.values[i:i+timesteps,:]
    return result

def to_lstm_output(output, timesteps):
    return output.values[:-timesteps + 1,:]
```

```python
x = to_lstm_input(ds.input_working, 3)
y = to_lstm_output(ds.output_working, 3)
```

```python
n_timesteps = np.shape(x)[1]
n_features = np.shape(x)[2]
output_dim = np.shape(y)[1]

model = keras.models.Sequential()
model.add(keras.layers.LSTM(50, activation='relu', return_sequences=True, input_shape=(n_timesteps, n_features)))
model.add(keras.layers.LSTM(50, activation='relu'))
model.add(keras.layers.Dense(output_dim))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x, y)
```
